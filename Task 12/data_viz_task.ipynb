```python
# Import libraries
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline # Graphs will be presented as a cell output.
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
pd.set_option('display.max_columns', None) # Allows for all of the columns to be displayed when calling the dataframe.
car_data.columns
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
selected_cars=car_data.loc[car_data['Manufacturer'].isin(selected_manufacturers)]
sns.boxplot(data=selected_cars, x='Rev.per.mile', y='Manufacturer')

plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Suzuki is the manufacturer with the highest revs per mile.


### 2. Histogram

```python
# Histogram of MPG
sns.histplot(x='MPG.city', data=car_data, bins=20, color='green', label='City')
sns.histplot(x='MPG.highway', data=car_data, bins=20, color='red', label='Highway')

plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.legend(title='Driving location')
plt.show()
# Two overlayed
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
It is generally more efficient to drive on the highway.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_data)
plt.xlabel('Wheelbase')
plt.ylabel('Turning circle')

plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The relationship is that as wheelbase gets larger so does the turning circle. 


### 4. Barplot

```python
# Barplot for mean horsepower
car_by_type = car_data.groupby(by=['Type'])
car_type_avg = round(car_by_type.mean(numeric_only=True), 0)# Creates a mean for each column that has a numerical columns.
 
sns.barplot(x='Type', y='Horsepower', data=car_type_avg, hue='Type', palette='muted')
plt.xlabel('Car type')
plt.ylabel('Horsepower')

plt.show()

```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
The answer is unclear.
The bar plot shows that large and midsize cars have more horsepower than small and compact vehicles, however the specific size of the vehicles is not a numerical value so a correlation of size against horsepower isn't particularily clear. A general assumption would be made that a van would be larger than a sporty 'size' car and possibly larger than a midsize car.
